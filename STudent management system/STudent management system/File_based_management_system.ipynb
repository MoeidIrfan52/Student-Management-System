```python
import numpy as np
import os

filename="students.txt"
if not os.path.exists(filename):
    open(filename,"w").close()

def get_student_id():
    while True:
        sid=input("Enter Student ID: ").strip()
        if sid.isdigit():
            return sid
        print("Error! Student ID must contain only numbers.")

def get_name():
    while True:
        name=input("Enter Name: ").strip()
        if name.replace(" ","").isalpha():
            return name
        print("Error! Name must contain only alphabets.")

def get_age():
    while True:
        age=input("Enter Age: ").strip()
        if age.isdigit() and int(age)>0:
            return age
        print("Error! Age must be a positive integer.")

def get_course():
    while True:
        c=input("Enter Course: ").strip()
        if c.replace(" ","").isalpha():
            return c
        print("Error! Course must contain only alphabets.")

def get_marks():
    while True:
        try:
            m=float(input("Enter Marks: "))
            if 0<=m<=100:
                return m
        except ValueError:
            pass
        print("Error! Marks must be numeric between 0 and 100.")

def read_records():
    with open(filename,"r") as f:
        return [i.strip().split(",") for i in f if i.strip()]

def write_records(records):
    with open(filename,"w") as f:
        for r in records:
            f.write(",".join(map(str,r))+"\n")

def add_student():
    records=read_records()
    sid=get_student_id()
    if any(r[0]==sid for r in records):
        print("Student ID Already Exists"); return
    records.append([sid,get_name(),get_age(),get_course(),get_marks()])
    write_records(records)
    print("Student Added Successfully")

def display_students():
    records=read_records()
    if not records:
        print("No Records Found"); return
    print("ID\tName\tAge\tCourse\tMarks")
    print("-"*50)
    for r in records:
        print("\t".join(map(str,r)))

def search_student():
    sid=get_student_id()
    for r in read_records():
        if r[0]==sid:
            print(f"ID: {r[0]}\nName: {r[1]}\nAge: {r[2]}\nCourse: {r[3]}\nMarks: {r[4]}")
            return
    print("Student Not Found")

def update_student():
    records=read_records(); sid=get_student_id()
    for i,r in enumerate(records):
        if r[0]==sid:
            records[i]=[sid,get_name(),get_age(),get_course(),get_marks()]
            write_records(records); print("Student Updated Successfully"); return
    print("Student Not Found")

def delete_student():
    records=read_records(); sid=get_student_id()
    new=[r for r in records if r[0]!=sid]
    if len(new)==len(records):
        print("Student Not Found")
    else:
        write_records(new); print("Student Deleted Successfully")

def average_marks():
    rec=read_records()
    if not rec: print("No Records Found"); return
    print("Average Marks =",round(np.mean([float(r[4]) for r in rec]),2))

def highest_marks():
    rec=read_records()
    if not rec: print("No Records Found"); return
    h=max(float(r[4]) for r in rec)
    for r in rec:
        if float(r[4])==h:
            print(r)

def lowest_marks():
    rec=read_records()
    if not rec: print("No Records Found"); return
    l=min(float(r[4]) for r in rec)
    for r in rec:
        if float(r[4])==l:
            print(r)

while True:
    print("\n===== STUDENT MANAGEMENT SYSTEM =====")
    print("1.Add Student\n2.Display Students\n3.Search Student\n4.Update Student\n5.Delete Student\n6.Average Marks\n7.Highest Marks\n8.Lowest Marks\n9.Exit")
    ch=input("Enter Choice: ").strip()
    if ch=="1": add_student()
    elif ch=="2": display_students()
    elif ch=="3": search_student()
    elif ch=="4": update_student()
    elif ch=="5": delete_student()
    elif ch=="6": average_marks()
    elif ch=="7": highest_marks()
    elif ch=="8": lowest_marks()
    elif ch=="9":
        print("Thank You!"); break
    else:
        print("Invalid Choice")
```
